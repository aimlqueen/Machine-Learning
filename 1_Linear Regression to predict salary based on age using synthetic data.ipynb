```python
# To avoid warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading the dataset
data=pd.read_csv("data/salary_synteticdata.csv")
data.head()
```

```python
# exploring the data
data.shape
```

```python
# data visualization :Boxplot
plt.boxplot(data['Salary'])
```

```python
plt.boxplot(data['Age'])
```

```python
# checking the relationship between Age & Salary
plt.plot(data['Age'],data['Salary'])
```

```python
# correlation check
data.corr()
```

```python
# splitting the data into X(input/independent variable/features) & Y(Target/Dependent variable/output)
# x:Age
# y : Salary
x=data['Age']
y=data['Salary']
```

```python
# reshape to 2D array/ dataframe
x=x.values.reshape(-1,1)
```

```python
x.shape
```

```python
y.head()
```

```python
# spliting data into train data & test data
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25) 
# 25% for test data
#  75% for traindata
```

```python
# linear regression modeling
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression() #creating object of Linearregression
```

```python
reg.fit(x_train,y_train) #training and fitting LR object using train data
```

```python
# model coefficients((m) 
reg.coef_
```

```python
 # intecept(c) of LR
reg.intercept_
```

```python
# predictions
ypred=reg.predict(x_test)
```

```python
ypred
```

```python
# calculate mean squaed error(MSE)
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(ypred,y_test)
MSE
```

```python
sum((y_test-ypred)**2)/len(ypred) #manually calculating MSE
```

```python
# creating a dataframe to view results
Results=pd.DataFrame(columns=['Acutal_Salary','Predicted_Salary','Age'])
Results['Acutal_Salary']=y_test
Results['Predicted_Salary']=ypred
Results['Age']=x_test
Results=Results.reset_index()
Results['Id']=Results.index
Results.head()
```

```python
# visualization of Actual vs predicated salary
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='Age',y='Acutal_Salary',data=Results)
plt.plot()
```

```python
# visualization of Actual vs predicated salary
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='Age',y='Predicted_Salary',data=Results)
plt.plot()
```

```python
# visualization of Actual vs predicated salary
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='Acutal_Salary',y='Predicted_Salary',data=Results)
plt.plot()
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,ypred)
r2
```
